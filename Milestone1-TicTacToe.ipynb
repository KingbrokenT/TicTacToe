```python
def tictactoe():
    print('Welcome to tic-tac-toe.')
    print('Here is what the board looks like each number corresponds to a position on the board')
    print('(   |   |   )\n( 1 | 2 | 3 )\n(___|___|___)\n(   |   |   )\n( 4 | 5 | 6 )\n(___|___|___)\n(   |   |   )\n( 7 | 8 | 9 )\n(   |   |   )')
    current_board = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    player = 'X'
    winner = []
    while winner == []:
        choice = select_position(player, current_board)
        
        new_board = replace_position(player,choice, current_board)
        
        winner = check_board(new_board, player)
        
        new_board = current_board
        
        showboard(current_board)
        if player == 'X':
            player = 'O'
        elif player == 'O':
            player = 'X'
    if winner == 'tie':
        print('There was a tie')
    else:
        print(f'There was a {winner}')
    
```

```python
def select_position(player, new_board):
    # this allows users to select a new position on the board
    position = []
    while position == []:
        pos = input(f'Select the  position for {player}: ')
        if pos.isdigit():
            if float(pos) in range(1,10):
                if new_board[int(pos)-1] == ' ':
                    position = int(pos)
                else:
                    print('Please select an empty position')
            else:
                print('Please select an interger between 1 and 9')
        else:
            print('Please select an interger between 1 and 9')
    return position         
```

```python
def replace_position(player,choice, current_board):
    current_board[choice - 1] = player
    new_board = current_board
    return new_board
```

```python
def check_board(new_board, player):
    if new_board[0] == new_board[1] == new_board[2] == player or new_board[3] == new_board[4] == new_board[5] == player or new_board[6] == new_board[7] == new_board[8] == player:
        winner = player
        return winner
    elif new_board[0] == new_board[3] == new_board[6] == player or new_board[1] == new_board[4] == new_board[7] == player or new_board[2] == new_board[5] == new_board[8] == player:
        winner = player
        return winner
    elif new_board[0] == new_board[4] == new_board[8] == player or new_board[2] == new_board[4] == new_board[6] == player:
        winner = player
        return winner
    for pos in new_board:
        if pos == ' ':
            winner = []
            return winner
    else:
        winner = 'tie'
        return winner
```
